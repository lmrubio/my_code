```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create a 50 element array with linearly spaced elements from 0 to 2pi
x = np.linspace(0, 2*np.pi, 50) 
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel("x label")
plt.ylabel("y label")
```
